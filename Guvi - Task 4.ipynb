# if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv

```python
exceeded_expectations = df[df['CodeKata Score']>15000]
exceeded_expectations.to_csv('exceeded_expectations.csv') # you can find the new exceeded_expectations.csv file in your local
exceeded_expectations
```

# if 10000<codekata score<15000 (Reached_expectations.csv)

```python
reached_expectations = df[(df['CodeKata Score']>10000) & (df['CodeKata Score']<15000)]
reached_expectations.to_csv('reached_expectations.csv') # you can find the new reached_expectations.csv file in your local
reached_expectations
```

# if 7000<codekata score<10000 (Needs_Improvement.csv)

```python
needs_improvement = df[(df['CodeKata Score']>7000) & (df['CodeKata Score']<10000)]
needs_improvement.to_csv('needs_improvement.csv') # you can find the new needs_improvement.csv file in your local
needs_improvement
```

# if codekate score < 7000 (Unsatisfactory.csv)

```python
unsatisfactory = df[df['CodeKata Score']<7000]
unsatisfactory.to_csv('unsatisfactory.csv') # you can find the new unsatisfactory.csv file in your local
unsatisfactory
```
